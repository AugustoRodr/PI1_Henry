```python
import numpy as np
import pandas as pd
from datetime import datetime
```

Carga de los datasets de las plataformas

```python
df_netflix= pd.read_csv('datasets/netflix_titles.csv').copy()
df_amazon= pd.read_csv('datasets/amazon_prime_titles.csv').copy()
df_disney= pd.read_csv('datasets/disney_plus_titles.csv').copy()
df_hulu= pd.read_csv('datasets/hulu_titles.csv').copy()

df_netflix['Plataforma'],df_amazon['Plataforma'],df_disney['Plataforma'],df_hulu['Plataforma']='netflix','amazon','disney','hulu'

df_total=pd.concat([df_netflix,df_amazon,df_disney,df_hulu], ignore_index=True)
df_total
```

Cambio los nulos del campo rating por 'G'

```python
df_total.rating.replace({np.nan:'G'}, inplace=True)
df_total.cast.replace({np.nan:'Sin Informacion'}, inplace=True)
df_total.duration.replace({np.nan:'0 min'}, inplace=True)
```

```python
#df_total.isna().sum()# Cuento la cantidad de datos faltantes en cada campo
df_total.info()# info general del dataframe
```

Genero el campo id

```python
df_total['id']=[df_total.Plataforma.iloc[i][0] + df_total.show_id.iloc[i] for i in range(len(df_total))]
df_total
```

- Cambio el formato del campo date_added

```python
df_total.sample(5)
```

```python
df_total.date_added=pd.to_datetime(df_total['date_added'], format='mixed')
```

- Campos de texto u object a minuscula

**nota: el siguiente codigo pone en minuscula todos los campos que sean object. Con esto evito las repeticiones innecesarias**

```python
lscampos=df_total.columns# lista con los nombres de las columnas

for campo in lscampos:
    if df_total[f'{campo}'].dtype==object:
        df_total[f'{campo}']=df_total[f'{campo}'].str.lower()
    else:
        pass
```

- El campo duration se convierte en 2 campos mÃ¡s

```python
df_total['duration_int']= df_total['duration'].apply( lambda row: str(row).split()[0])
df_total['duration_type']= df_total['duration'].apply( lambda row: str(row).split()[1])
```

```python
df_total.sample(5)
```

## Unifico el dataframe de ratings

```python
df1=pd.read_csv('datasets/ratings/1.csv').copy()
df2=pd.read_csv('datasets/ratings/2.csv').copy()
df3=pd.read_csv('datasets/ratings/3.csv').copy()
df4=pd.read_csv('datasets/ratings/4.csv').copy()
df5=pd.read_csv('datasets/ratings/5.csv').copy()
df6=pd.read_csv('datasets/ratings/6.csv').copy()
df7=pd.read_csv('datasets/ratings/7.csv').copy()
df8=pd.read_csv('datasets/ratings/8.csv').copy()
df_ratings=pd.concat([df1,df2,df3,df4,df5,df6,df7,df8], ignore_index=True)
```

```python
df_ratings.info()
#df_ratings.isna().sum()
```

```python
df_ratings
```

- calculo el promedio del df de ratings

```python
ratings_prom=df_ratings.groupby('movieId').rating.agg([sum,len])
```

```python
ratings_prom.head()
```

```python
ratings_prom['score']=ratings_prom['sum']/ratings_prom['len']
```

```python
ratings_prom.reset_index(inplace=True)
```

- Hago un join de la tabla de ratings hacia la tabla de plataformas

```python
df_total=df_total.join(ratings_prom.set_index('movieId'), on='id', lsuffix='_caller', rsuffix='_other')
```

```python
df_total
```

```python
df_total.to_csv('df_general.csv', index=False)
```

```python
df_testeo=pd.read_csv('df_general.csv').copy()
```

```python
# CREACION, CARGA Y CONTEO DE ACTORES SEGUN LOS FILTROS
lista_anio=sorted(list(df_testeo.release_year.unique()))
lista_plat=list(df_testeo.Plataforma.unique())
act_dict={plat:{anio:{} for anio in lista_anio} for plat in lista_plat}
for plat in lista_plat:
    df_temp=df_testeo[df_testeo.Plataforma==plat]
    for anio in lista_anio:
        df_temp2=df_temp[df_temp.release_year==anio]
        for i in df_temp2.cast:
            if i != 'sin informacion':
                actores=i.split(',')
                for actor in actores:
                    actor=actor.strip()
                    if actor in act_dict[plat][anio]:
                        act_dict[plat][anio][actor]+=1
                    else:
                        act_dict[plat][anio][actor]=1
```

```python
tabla_actores=[]
for plataforma in act_dict:
    print(plataforma)
    for anios in act_dict[plataforma]:
        for actor in act_dict[plataforma][anios]:
            tabla_actores.append([plataforma,anios,actor,act_dict[plataforma][anios][actor]])
```

```python
df_query4=pd.DataFrame(tabla_actores, columns=['Plataforma','Anios','Actor','Cantidad'])
df_query4.iloc[95984].Actor
```
